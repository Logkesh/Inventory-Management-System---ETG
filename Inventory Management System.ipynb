```python
# CODE FOR INVENTORY MANAGEMENT SYSTEM

ch = ''

print('*'*20,' Welcome to ETG Super Market ','*'*20)

while ch != '6':
    
    with open('IMS.json' ,'r+') as file:
            data = json.loads(file.read())                   # LOADING THE DICTIONARY TO A VARIABLE FROM THE FILE
            
    ch = input('\nMAIN MENU' + 
           '\n\t1) View Items' + 
           '\n\t2) Add New Item' +                           # MAIN MENU
           '\n\t3) Buy An Item' +
           '\n\t4) Update An Item' +
           '\n\t5) Delete An Item' +
           '\n\t6) Exit' +
           '\nEnter (1/2/3/4/5/6) to select the options : ')
    
    
    if ch == '1':                                            # WHEN THE USER WANTS TO VIEW THE DETAILS OF ALL STOCKS
            for i in data:
                print('\nName : ', data[i]['name'] ,
                      '\tPrice : ', data[i]['price'] ,
                      '\tQuantity : ', data[i]['quantity'] ,
                      '\tCategory : ', data[i]['category'] ,
                      '\tNet Weight : ', data[i]['net weight'])
                
                
    elif ch == '2':                                          # WHEN THE USER NEEDS TO ADD AN ITEM
            name = input("\nEnter the New Item Name : ")
            price = float(input("Enter the New Item Price : "))
            quan = int(input("Enter the New Item Quantity : "))
            cat = input("Enter the New Item Category \n(chocolate/Snack/Juice/Nutrition Drink/Biscuit) : ")
            nweit = str(input("Entre the New Item Net Weight in grams(g)/liters(l) : "))
            
            pid = 20210900 + (len(data)+1)
            
            data.update({pid:{'name':name,'price':price,'quantity':quan,'category':cat,'net weight':nweit}})
            
            f = open('IMS.json','w')
            f.write(json.dumps(data))                # UPDATING THE JSON FILE
            f.close()
            
            print("\nNew Item Added...")
            
            
    elif ch == '3':                                         # WHEN THE COSTUMER NEEDS TO PURCHASE ANY ITEM
        count = 0
        pur = '1'
        plist = []
        tprice = 0.0
        for i in data:                                     # DISPLAY'S THE LIST OF PRODUCTS WITH THE PRICE AND PRODUCT ID
            print("\n",data[i]['name'] , " -> ","Cost : ", data[i]['price'] ," -> ", list(data.keys())[count])
            count+=1
            
        while(pur == '1'):                                 # WHILE LOOP IS FOR MULTIPLE PURCHASING
            name = ''
            while (name not in list(data.keys())):
                name = input("\nCopy & Paste the Id of the Product here... \nto Buy the product from the above list : ")
                                                           # GETS THE PRODUCT ID FOR ACCESSING THE PRODUCT DETAILS
                if name not in list(data.keys()):
                    print("\nInvalid name...Try Again ")
                    
            while True:
                quant = int(input("\nTotally " + str(data[name]['quantity']) + " stocks left"+
                          "\nEnter the Quantity of the product... \nYou need to buy : "))
                
                if quant>int(data[name]['quantity']):
                    print("\nThe Entered quantity is more than the stock... \nEnter the correct stock... ")
                    
                else : break
            tprice += quant * data[name]['price']
            print("Total Cost : ",quant * data[name]['price'])
            data[name]['quantity']-=quant
            
            plist.append(data[name]['name'] + " - " +str(quant))    # APPENDING THE DATA INTO THE LIST
            
            pur = input("\nEnter '1' to puchase 1 more item else give 'enter' key to exit : ")
            
        print("\nProcessing the purchase...")
        
        f = open('IMS.json','w')
        f.write(json.dumps(data))                # UPDATING THE JSON FILE
        f.close()
        
        print("\nPurchase Done...")
        
        print("\nPurchased Items : \n")
        count = 0
        for i in plist:                           # PRINTING THE PURCHASED ITEMS
            count+=1
            print(count, ". ",i)
        print("\nTotal Amount : ", tprice)
            
        print("\nThank's For Purchasing...\n Come Again... :)")
        
        
    elif ch == '4':                                        # WHEN THE USER NEEDS TO UPDATE ANY DETAIL A SPECIFIC ITEM
        count = 0
        
        for i in data:                                     # DISPLAY'S THE LIST OF PRODUCTS WITH PRODUCT ID
            print("\n",data[i]['name'] , "-> ", list(data.keys())[count])
            count+=1
            
        pid = ''
        while pid not in list(data.keys()):
            pid = input("\nCopy & Paste the Id of the Product here... \nto update the product from the above list : ")
                                                           # GETS THE PRODUCT ID FOR ACCESSING THE PRODUCT DETAILS
            if pid not in list(data.keys()):
                print("\nInvalid id...Try Again ")
                
        print("\nProduct Detail : ")
        print('\nName : ', data[pid]['name'] ,
                      '\nPrice : ', data[pid]['price'] ,
                      '\nQuantity : ', data[pid]['quantity'] ,
                      '\nCategory : ', data[pid]['category'] ,
                      '\tNet Weight : ', data[i]['net weight'])
        
        att = ''
        while att not in ['name','price','quantity','category','net weight']:
            att = input('\nEnter the attribute to be updated \n(name/price/quantity/category/net weight) :')
            att = att.lower()
            if att not in ['name','price','quantity','category','net weight']:
                print("\nINVALID ATTRIBUTE ... TRY AGAIN...")
                
        data[pid][att] = type(data[pid][att])(input('Enter the new value : '))          # GETS THE NEW DETAIL FROM THE USER...
                                                                                        # AND THE UPDATES THE DATA
        print("\nProcessing the update...")
        
        f = open('IMS.json','w')
        f.write(json.dumps(data))                # UPDATING THE JSON FILE
        f.close()
        
        print("Update Completed...")
                    
    elif ch == '5':                                        # WHEN THE USER NEEDS TO DELETE ANY DETAILS
        count = 0
        
        for i in data:                                     # DISPLAY'S THE LIST OF PRODUCTS WITH PRODUCT ID
            print("\n",data[i]['name'] , "-> ", list(data.keys())[count])
            count+=1
            
        pid =''
        while pid not in list(data.keys()):
                pid = input("\nCopy & Paste the Id of the Product here... \nto delete the product from the above list : ")
                                                           # GETS THE PRODUCT ID FOR ACCESSING THE PRODUCT DETAILS
                if pid not in list(data.keys()):
                    print("\nInvalid ID...Try Again ")
                    
        del data[pid]                            # DELETING THE PRODUCT DETAILS
        
        f = open('IMS.json','w')
        f.write(json.dumps(data))                # UPDATING THE JSON FILE
        f.close()
        
        print("\nItem Deleted Successfully...")
        
        
    elif ch == '6':                              # EXITING THE PROGRAM
        print("\nExiting...")
        print("Thank's For using Me ;)")
        
        
    else:
        print("\nInvalid Input... Try Again")      # FOR INVALID INPUT
```
